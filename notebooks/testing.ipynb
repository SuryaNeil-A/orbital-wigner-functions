```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
def eigenvals(k, m, t):
    val = np.sqrt(m**2 + t**2 + 2*m*t*np.cos(k))
    return val, -1*val
```

```python
k = np.linspace(0, 2*np.pi, 200)

fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(k, eigenvals(k, 0, 1)[0], label=r"$m=0,\ \tau=1$ Band 1", color="tab:blue")
ax.plot(k, eigenvals(k, 0, 1)[1], label=r"$m=0,\ \tau=1$ Band 2", color="tab:green")

ax.plot(k, eigenvals(k, 1, 1)[0], label=r"$m=1,\ \tau=1$ Band 1", color="tab:red")
ax.plot(k, eigenvals(k, 1, 1)[1], label=r"$m=1,\ \tau=1$ Band 2", color="tab:orange")

ax.plot(k, eigenvals(k, 0, 2)[0], label=r"$m=0,\ \tau=2$ Band 1", color="tab:purple")
ax.plot(k, eigenvals(k, 0, 2)[1], label=r"$m=0,\ \tau=2$ Band 2", color="tab:brown")

ax.set_title("Band Structure Plot")
ax.grid()
ax.legend()

ax.set_xlabel("k")
ax.set_ylabel("E")

ax.set_xlim(0, 2*np.pi)

plt.show()
```
